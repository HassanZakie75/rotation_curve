Modified Bessel Function of the First Kind (I_n(x))

The scipy.special.iv(n, x) function calculates the modified Bessel function of the first kind of order n.

Let's plot I_0(x) and I_1(x) for a range of x values.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import iv

# Generate x values
x = np.linspace(0, 5, 100)

# Calculate I_0(x) and I_1(x)
i0 = iv(0, x)
i1 = iv(1, x)

# Plot the functions
plt.figure(figsize=(10, 6))
plt.plot(x, i0, label='$I_0(x)$', color='blue')
plt.plot(x, i1, label='$I_1(x)$', color='red')

# Add labels and title
plt.xlabel('x', fontsize=14)
plt.ylabel('Value', fontsize=14)
plt.title('Modified Bessel Functions of the First Kind', fontsize=16)
plt.legend(fontsize=12)
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.show()
```

Modified Bessel Function of the Second Kind (K_n(x))

The scipy.special.kv(n, x) function calculates the modified Bessel function of the second kind of order n.

Let's plot K_0(x) and K_1(x) for a range of x values. Note that K_n(x) is singular at x=0, so we should start our x range slightly greater than 0 to avoid a division-by-zero error.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import kv

# Generate x values, starting slightly above 0
x = np.linspace(0.1, 5, 100)

# Calculate K_0(x) and K_1(x)
k0 = kv(0, x)
k1 = kv(1, x)

# Plot the functions
plt.figure(figsize=(10, 6))
plt.plot(x, k0, label='$K_0(x)$', color='green')
plt.plot(x, k1, label='$K_1(x)$', color='purple')

# Add labels and title
plt.xlabel('x', fontsize=14)
plt.ylabel('Value', fontsize=14)
plt.title('Modified Bessel Functions of the Second Kind', fontsize=16)
plt.legend(fontsize=12)
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.show()
```

Combining Modified Bessel Functions

Often, the solution to a physical problem involves a linear combination of I_n(x) and K_n(x). For example, a common solution for problems with cylindrical symmetry is of the form:

where A and B are constants determined by boundary conditions, and 
alpha is a constant related to the physical properties of the system.

Let's plot a combined function to see how they behave together. Consider a simple case where we add I_0(x) and K_0(x).

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import iv, kv

# Generate x values (starting above 0)
x = np.linspace(0.1, 5, 100)

# Calculate I_0(x) and K_0(x)
i0 = iv(0, x)
k0 = kv(0, x)

# Combine the functions
combined_function = i0 + k0

# Plot the individual and combined functions
plt.figure(figsize=(12, 8))
plt.plot(x, i0, label='$I_0(x)$', color='blue', linestyle='--')
plt.plot(x, k0, label='$K_0(x)$', color='green', linestyle='--')
plt.plot(x, combined_function, label='$I_0(x) + K_0(x)$', color='red', linewidth=2)

# Add labels and title
plt.xlabel('x', fontsize=14)
plt.ylabel('Value', fontsize=14)
plt.title('Combination of Modified Bessel Functions', fontsize=16)
plt.legend(fontsize=12)
plt.grid(True)
plt.show()
```
